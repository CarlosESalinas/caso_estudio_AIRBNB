## Tablas de frecuencia 

Para la solución del caso de uso es necesario identificar una posible hipótesis para validar a través de las herramientas que se revisarán en las siguientes líneas.

La tabla de frecuencia es una herramienta que permite ordenar los datos que se presentan numéricamente las características de la distribución de un conjunto de datos o muestra. 

**Hipótesis:** la categoría de celulares es la que más productos vende, se quiere validar ya que se desea lanzar una campaña de descuento para traer más clientes a la empresa, sin embargo, se debe conocer cuál va a ser la categoría en la cual se invertirá más publicidad. 

```python
data.columns
```

```python
## Tabla de frecuencia univariada 
# Se va a realizar una tabla para analizar las categorías que se venden en el comercio. 
# Frecuencia absoluta

frec_data = data['nombre_categoria_producto'].value_counts().reset_index()
frec_data.head()
```

```python
# Frecuencia relativa 
frec_data['porcentaje_nombre_categoria_producto'] = frec_data['nombre_categoria_producto'].apply(lambda x: x/frec_data['nombre_categoria_producto'].sum())
frec_data.head(5)
```

```python
# Tabla de frecuencia absoluta y relativa acumulada con comsum()
## Frecuencia absoluta acumulada 
frec_data['acumulada_nombre_categoria_producto']=frec_data['nombre_categoria_producto'].cumsum()

## Frecuencia relativa acumulada 
frec_data['acumulada_porcentaje_nombre_categoria_producto']=frec_data['porcentaje_nombre_categoria_producto'].cumsum()
frec_data.head()
```

```python
frec_data
```

*Análisis**: La tabla de frecuencias nos muestra la siguiente información:
1. Frecuencia absoluta: número de veces que aparece un valor.
2. Frecuencia relativa: el resultado de dividir la frecuencia absoluta de un determinado valor por el número total de datos. 
3. Frecuencia acumulada: la suma de las frecuencias absolutas de todos los valores iguales o menores al valor considerado.
4. Frecuencia acumulada relativa: el resultado de dividir la frecuencia acumulada por el número de toda la información. 

```python
## Tablas de frecuencia bivariada
# Tabla que muestra la frecuencia por ciudad de venta y categoría del producto
pd.crosstab(index=data['nombre_categoria_producto'], columns=data['ciudad_vendedor'])
```

**Análisis**: Es posible en la tabla ver la cantidad de compras que tiene las diferentes categorías, la que más tiene es *Tecnología* con 532 ventas en total. 

## Tendencia central 

La tendencia central significa medir el centro o distribución de ubicación de valores de un conjuto de datos. Da una idea del valor promedio de las variables en el conjunto de datos y también una indicación de cuán ampliamente se distribuyen los valore en el conjunto de datos.

1. Media
2. Mediana
3. Moda

**Hipótesis**: Los precios en el comercio son mayores a 500, ya que el estudio arrojó que los clientes más recurrentes compran mayormente productos más ecnonómicos (debajo de 500), por tanto, es necesario validar si en promedio la compañía tiene precios más competitivos.

```python
## Media o Promedio 
# Promedio de precios
round(data['precio'].mean(),2)
```

```python
## Mediana
# Mediana de precio, valor central de la distribución de los datos.
round(data['precio'].median(),2)
```

```python
## Moda
# Valor más común de los precios
data['precio'].mode()
```

**Análisis**: Es posible analizar que el promedio es inferior a 500 ubicándose en 120, la mediana que es el dato central se ubica en 83.12 y existen varios valores comunes y todos debajo de 60 lo que nos muestra que la mayor cantidad de precios son bajos y que permiten ser más atractivos para el consumidor final que compra productos mayormente debajo de 500.

## Medidas de localización 

Los *cuantiles* y *percentiles* son puntos de corte que dividen el rango de una distribución de probabilidad en intervalos continuos son probabilidades iguales o que dividen a las observaciones en una muestra de la misma forma.

Así se puede representar un porcentaje de datos a través de los cuantiles para indetificar cómo están dividos los datos. 

**Hipótesis**: la mitat (50%) de los productos pesa más de 500 gramos, ya que la empresa de mensajería quiere realizar un descuento si los productos pesan menos de 500 gramos, así la compañía podría bajar los precios de envío de la mitad de los producto que se venden.

```python
## Cuantiles 
# Calcular el 20%, 50% y 70% del peso de los productos. 
data['peso_g_producto'].quantile([0.20,0.50,0.70])
```

**Análisis**: El percentil 50 tiene un valor de 1458.5 lo cual nos indica que la mitad de los productos tiene un peso de más alto que 500 gramos, por lo que se rechaza la hipótesis y nos indica que no se puede realizar el descuento sobre el vólumen del producto, sin embargo deja abierta la puerta para encontrar qué cantidad de producto pueden contar con el descuento. 

## Medidas de dispersión 

Se utilizan las estadísticas de dispersión para dar un número único que describa qué tan compacto o disperso es un conjunto de observaciones. Estas se cumplementan de las medidas mencionadas anteriormente. 

Estas medidas entre _más grandes_, representan más dispersión y entre más pequeñas más concentradas están las observaciones (es importante tener en cuenta la escala de la variable para analizar la dispersión).

1. Desviación estándar
2. Rango intercuartílico
3. Rango 
4.  Coeficiente de variación

**Hipótesis**: La variación de las longitudes de los datos es mayor a 20cm, ya que las empresas de mensajería que tienen cubículos de transporte de mensajería individual de producto de longitud de 20 cm son más económicas, y así se podría cambiar estas empresas y ahorrar costos de envío . 

```python
## Desviación estándar 
# Encontrar la desviación estándar de la longitud de los productos de la muestra
# Represental qué tan lejos de la media se encuentran las observaciones 
round(data['longitud_cm_producto'].std(),2)
```

```python
## Rango intercuartílico 
# Encontra el rango intercuartílico de la longitud de los productos
# es la diferencia entre el tercer y primer cuartil. 
round(data['longitud_cm_producto'].quantile(0.75)-data['longitud_cm_producto'].quantile(0.25),2)
```

```python
# Rango 
# Encontrar el rango de la longitud de los productos de la muestra 
# Representa la diferencia que existe entre el producto más y el más pequeño
data['longitud_cm_producto'].max()-data['longitud_cm_producto'].min()
```

```python
## Coeficiente de variación 
# Encontrar el coeficiente de variación de la longitud de los productos. 
# Es la división de la desviación estándar sobre la media
round(data['longitud_cm_producto'].std()/data['longitud_cm_producto'].mean(),3)
```

**Análisis**: a pesar de que la desviación estándar y el rango intercuartílico son menores a 25, se puede alcanzar con el rango que la variación entre el máximo y el mínimo es de 91 cm, por tanto se podría migrar alredador del 50% de los productos de mensajería con el cubículo de 20 cm y darle al cliente un descuento en el envío.
